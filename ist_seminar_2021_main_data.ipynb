```python
import bdpy
```

```python
bdata = bdpy.BData('data/YSH210702_IstArtists_volume_native_prep.h5')
#bdata = bdpy.BData('data/KMT210702_IstColor_volume_native_prep_train.h5')
#bdata = bdpy.BData('KMT210702_IstColor_volume_native_prep_test.h5')
```

### ROI (brain region)

```python
# V1
brain_data_V1 = bdata.select('ROI_V1')
print(brain_data_V1.shape)

# LOC
brain_data_LOC = bdata.select('ROI_LOC')
print(brain_data_LOC.shape)

# FEF (HCP-180 parcellation; Glasser et al., 2016)
brain_data_FEF = bdata.select('hcp180_*_FEF_ROI')
print(brain_data_FEF.shape)

# The whole cortex
brain_data_cortex = bdata.select('VoxelData')
print(brain_data_cortex.shape)
```

Available ROIs:

- Eary visual areas (HCP180 based)
    - V1: `ROI_V1`
    - V2: `ROI_V2`
    - V3: `ROI_V3`
    - V4: `ROI_hV4`
- Ventral visual areas (functionally defined)
   - LOC: `ROI_LOC`
   - FFA: `ROI_FFA`
   - PPA: `ROI_PPA`
- HCP 180 parcellations
   - ROI list: <https://gist.github.com/ShuntaroAoki/b61791ec72cfcbecf1d7df89fcde90ea>
   - See also [Glasser et al., 2016](http://dx.doi.org/10.1038/nature18933)
